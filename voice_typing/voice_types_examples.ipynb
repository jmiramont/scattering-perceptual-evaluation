```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('signals_saar.csv', header=None)
data = data.to_numpy().T
data.shape
```

```python
#The axes size is determined by the figure size and the figure spacings, which can be set using figure.subplots_adjust(). In reverse this means that you can set the axes size by setting the figure size taking into acount the figure spacings:

import matplotlib.pyplot as plt

def set_size(w,h, ax=None):
    """ w, h: width, height in inches """
    if not ax: ax=plt.gca()
    l = ax.figure.subplotpars.left
    r = ax.figure.subplotpars.right
    t = ax.figure.subplotpars.top
    b = ax.figure.subplotpars.bottom
    figw = float(w)/(r-l)
    figh = float(h)/(t-b)
    ax.figure.set_size_inches(figw, figh)
```

```python
fig, axs = plt.subplots(3,1 )
# Type 1
axs[0].plot(data[27,9000:10501],linewidth = 1) #6
axs[0].set_xlim([0,1501])
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].set_title('Type 1', fontsize = 9)
axs[0].set_ylabel('amplitude')
# Type 2
axs[1].plot(data[15,9000:11001],linewidth = 1) # 88
axs[1].set_xlim([0,2001])
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].set_title('Type 2', fontsize = 9)
axs[1].set_ylabel('amplitude')
# Type 3
axs[2].plot(data[0,9000:11001],linewidth = 1)
axs[2].set_xlim([0,2001])
axs[2].set_xticks([])
axs[2].set_yticks([])
axs[2].set_title('Type 3', fontsize = 9)
axs[2].set_xlabel('time')
axs[2].set_ylabel('amplitude')

# fig.set_size_inches((6,5))
set_size(6,3)
plt.savefig('figures/voice_types_examples/voice_types_examples.pdf',bbox_inches='tight')
```
